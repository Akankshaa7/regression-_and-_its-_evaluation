Question 1: What is Simple Linear Regression?

--->**Simple Linear Regression** is a **statistical method** used to study the relationship between **two variables** ‚Äî one **independent variable (X)** and one **dependent variable (Y)**.

It helps us understand how a change in the independent variable affects the dependent variable by fitting a **straight line** (called the **regression line**) to the data points.

The relationship is represented by the equation:

[
Y = b_0 + b_1X + \epsilon
]

Where:

* **Y** ‚Üí Dependent variable (the outcome we want to predict)
* **X** ‚Üí Independent variable (the input or predictor)
* **b‚ÇÄ** ‚Üí Intercept (value of Y when X = 0)
* **b‚ÇÅ** ‚Üí Slope (rate of change in Y for one unit change in X)
* **Œµ** ‚Üí Error term (difference between predicted and actual values)

**Example:**
Suppose we want to predict a person‚Äôs **weight (Y)** based on their **height (X)**. Simple linear regression will find the best-fit line that shows how weight changes with height.




Question 2: What are the key assumptions of Simple Linear Regression?

--->The **key assumptions of Simple Linear Regression** ensure that the model‚Äôs predictions and interpretations are valid. There are **five main assumptions**, described below:

1. **Linearity** ‚Äì
   There must be a **linear relationship** between the independent variable (X) and the dependent variable (Y).
   ‚û§ Example: As X increases, Y should increase or decrease at a constant rate.

2. **Independence of Errors** ‚Äì
   The **residuals (errors)** ‚Äî the differences between actual and predicted Y values ‚Äî should be **independent** of each other.
   ‚û§ This means one observation‚Äôs error should not influence another‚Äôs.

3. **Homoscedasticity** ‚Äì
   The **variance of residuals** should be **constant** across all levels of X.
   ‚û§ If residuals spread out more for higher or lower X values, it indicates **heteroscedasticity**, which violates this assumption.

4. **Normality of Errors** ‚Äì
   The **residuals** should be **normally distributed** (bell-shaped curve) around zero.
   ‚û§ This is important for valid hypothesis tests and confidence intervals.

5. **No Multicollinearity (for multiple regression)** ‚Äì
   In **simple** linear regression, there is only one X variable, so this assumption is automatically satisfied.
   ‚û§ It applies mainly when there are multiple predictors.




Question 3: What is heteroscedasticity, and why is it important to address in regression models?

--->**Heteroscedasticity** refers to a situation in regression analysis where the **variance of the residuals (errors)** is **not constant** across all levels of the independent variable(s).

In simpler terms, it means that as the value of X changes, the **spread (or scatter)** of the errors around the regression line **changes** ‚Äî instead of remaining roughly the same.

---

### üîπ Example:

If you‚Äôre predicting **income (Y)** based on **years of education (X)**, you might notice that the prediction errors are small for people with fewer years of education but much larger for people with higher education levels.
This shows **heteroscedasticity** ‚Äî the variance of errors increases with X.

---

### üîπ Why It‚Äôs Important:

Heteroscedasticity **violates a key assumption** of linear regression ‚Äî that residuals have **constant variance (homoscedasticity)**.
If not addressed, it can cause several problems:

1. **Biased standard errors** ‚Äì The estimates of the standard errors become unreliable.
2. **Invalid hypothesis tests** ‚Äì t-tests and F-tests may give misleading results.
3. **Incorrect confidence intervals and p-values** ‚Äì They may be too narrow or too wide, leading to wrong conclusions.
4. **Reduced prediction accuracy** ‚Äì The model‚Äôs predictions may become less reliable for certain ranges of X.





Question 4: What is Multiple Linear Regression?

--->**Multiple Linear Regression (MLR)** is a statistical technique used to study the relationship between **one dependent variable (Y)** and **two or more independent variables (X‚ÇÅ, X‚ÇÇ, X‚ÇÉ, ‚Ä¶)**.

It extends **Simple Linear Regression**, which uses only one predictor, by allowing us to analyze and predict outcomes based on **several factors simultaneously**.

---

### üîπ The Equation:

[
Y = b_0 + b_1X_1 + b_2X_2 + b_3X_3 + ... + b_nX_n + \epsilon
]

Where:

* **Y** ‚Üí Dependent variable (the outcome to be predicted)
* **X‚ÇÅ, X‚ÇÇ, X‚ÇÉ, ‚Ä¶ X‚Çô** ‚Üí Independent (predictor) variables
* **b‚ÇÄ** ‚Üí Intercept (value of Y when all X = 0)
* **b‚ÇÅ, b‚ÇÇ, b‚ÇÉ, ‚Ä¶ b‚Çô** ‚Üí Regression coefficients (show how much Y changes for a one-unit change in each X, keeping others constant)
* **Œµ** ‚Üí Error term (difference between actual and predicted values)

---

### üîπ Example:

Suppose you want to predict a **student‚Äôs final exam score (Y)** based on:

* **Hours studied (X‚ÇÅ)**
* **Number of assignments completed (X‚ÇÇ)**
* **Attendance percentage (X‚ÇÉ)**

Then the equation might look like:
[
\text{Exam Score} = 25 + 3.2X_1 + 1.5X_2 + 0.4X_3
]

This means:

* For every extra hour studied, the score increases by 3.2 points (holding X‚ÇÇ and X‚ÇÉ constant).
* For each additional assignment completed, the score increases by 1.5 points, etc.




Question 5: What is polynomial regression, and how does it differ from linear
regression?

--->Polynomial Regression is a type of regression analysis that models the relationship between the independent variable (X) and the dependent variable (Y) as an nth-degree polynomial.

It is an extension of linear regression, which fits a straight line, while polynomial regression can fit curved (non-linear) relationships.



The main difference between **linear regression** and **polynomial regression** lies in the type of relationship they model between the independent variable (X) and the dependent variable (Y).

In **linear regression**, the relationship between X and Y is assumed to be **straight-line** or **linear**. This means that as X increases or decreases, Y changes at a **constant rate**. The equation of a linear regression model is ( Y = b_0 + b_1X + \epsilon ), where the highest power of X is 1. Linear regression is simple, easy to interpret, and works well when the data points form a roughly straight pattern.

On the other hand, **polynomial regression** is used when the relationship between X and Y is **non-linear** or **curved**. It extends linear regression by adding **higher-degree terms** of X (like ( X^2, X^3, ) etc.) to capture curvature in the data. Its general equation is ( Y = b_0 + b_1X + b_2X^2 + b_3X^3 + ... + b_nX^n + \epsilon ). Polynomial regression provides more flexibility to fit complex patterns but can also lead to **overfitting** if the degree of the polynomial is too high.




Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = [1, 2, 3, 4, 5]
Y = [2.1, 4.3, 6.1, 7.9, 10.2]

X = np.array(X).reshape(-1, 1)
Y = np.array(Y)
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, label='Original Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

‚óè Area = [1200, 1500, 1800, 2000]
‚óè Rooms = [2, 3, 3, 4]
‚óè Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results

## Import libraries

### Subtask:
Import the necessary libraries, including `numpy` and `pandas` for data handling, `sklearn.linear_model` for the linear regression model, and `statsmodels.stats.outliers_influence` for calculating VIF.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
data = {'Area': [1200, 1500, 1800, 2000],
        'Rooms': [2, 3, 3, 4],
        'Price': [250000, 300000, 320000, 370000]}

df = pd.DataFrame(data)

X = df[['Area', 'Rooms']]
Y = df['Price']

display(df)
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
X = df[['Area', 'Rooms']]

vif_data = pd.DataFrame()
vif_data['variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

display(vif_data)
```

```python
print("VIF Results:")
print(vif_data)

print("\nExplanation of Multicollinearity:")
print("The VIF values for both 'Area' and 'Rooms' are approximately 127.80.")
print("A general threshold for indicating significant multicollinearity is a VIF value greater than 5 or 10.")
print("Since both variables have VIF values significantly higher than this threshold, it indicates strong multicollinearity between 'Area' and 'Rooms'.")
print("This means that 'Area' and 'Rooms' are highly correlated with each other.")
print("High multicollinearity can make it difficult to interpret the individual coefficients of the related variables in the regression model and can lead to unstable coefficient estimates.")
```

Question 8: Implement polynomial regression on the following data:
‚óè X = [1, 2, 3, 4, 5]

‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
X = [1, 2, 3, 4, 5]
Y = [2.2, 4.8, 7.5, 11.2, 14.7]

X_np = np.array(X).reshape(-1, 1)
Y_np = np.array(Y)
```

```python
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X_np)
```

```python
model = LinearRegression()
model.fit(X_poly, Y_np)
```

```python
Y_pred = model.predict(X_poly)
```

```python
plt.scatter(X_np, Y_np, label='Original Data')
plt.plot(X_np, Y_pred, color='red', label='Polynomial Regression Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data:
‚óè X = [10, 20, 30, 40, 50]
‚óè Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.

**Reasoning**:
Import the required libraries for data manipulation, plotting, and linear regression modeling.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = [10, 20, 30, 40, 50]
Y = [15, 35, 40, 50, 65]

X_np = np.array(X).reshape(-1, 1)
Y_np = np.array(Y)
```

```python
model = LinearRegression()
model.fit(X_np, Y_np)
```

```python
Y_pred = model.predict(X_np)
residuals = Y_np - Y_pred

plt.scatter(X_np, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()

print("Assessment of Heteroscedasticity:")
print("Examine the residuals plot. If the spread of the residuals is roughly constant across all values of the independent variable (X), it suggests homoscedasticity (constant variance of errors).")
print("If the spread of the residuals increases or decreases as X changes, it suggests heteroscedasticity (non-constant variance of errors).")
```

Another option is to use **regularization methods** like **Ridge** or **Lasso regression**, which can reduce the impact of multicollinearity by penalizing large coefficients.

Finally, after making these changes, I would **rebuild and test the model** again to make sure the issues are resolved and that the predictions are stable and accurate.
